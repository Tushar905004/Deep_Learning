```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
random_data=np.arange(-5,10,0.1)
```

```python
def visualization(function,yaxis =(-2.0,2.0)):
    ylim = (yaxis)
    plt.locator_params(nbins = 5)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.axhline(lw=1,c='black')
    plt.axvline(lw=1,c='black')
    plt.grid(alpha= 0.4,ls = '-')
    plt.box(on=None)
    plt.plot(random_data,function(random_data),c='r',lw=3)
    plt.show()
```

```python
def sigmoid(data):
    return 1/(1+np.exp(-data))
```

```python
visualization(sigmoid,yaxis=(-0.5,0.5))
```

```python
Relu_activation_function=np.vectorize(lambda random_data: random_data if random_data>0 else 0)
```

```python
visualization(Relu_activation_function, yaxis = (-3.0,3.0))
```

```python
def softmax(rand):
    return np.exp(rand)/np.sum(np.exp(rand),axis=0)
```

```python
rand=random_data
plt.plot(rand,softmax(rand))
plt.locator_params(nbins = 5)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.axhline(lw=1,c='black')
plt.axvline(lw=1,c='black')
plt.grid(alpha= 0.4,ls = '-')
plt.box(on=None)
plt.plot(rand,softmax(rand),c='r',lw=3)
plt.title('Activation Function : Softmax')
plt.show()
```
